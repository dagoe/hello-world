```
from ipywidgets import AppLayout, Button, Layout, GridspecLayout
from ipywidgets import Button, Layout, jslink, IntText, IntSlider

def create_expanded_button(description, button_style):
    return Button(description=description, button_style=button_style, layout=Layout(height='auto', width='auto'))

def create_colored_button(desc, col):
  b = Button(description=desc, layout=Layout(width='auto',height='80px'))
  b.style.button_color = col 
  return b

grid = GridspecLayout(5, 5)

for i in range(5):
    for j in range(5):
        grid[i, j] = create_colored_button(' {} - {}'.format(i, j), 'lightgreen')
grid
```
